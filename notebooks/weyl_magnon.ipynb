# Weyl Magnon Antiferromagnet 

In this notebook, I present the python code for one of my research papers [Phys. Rev. B 97, 094412 (2018)](https://doi.org/10.1103/PhysRevB.97.094412). This paper proposed the first existence of Weyl magnon in stacked kagome antiferromagnets, governed by the Heisenberg Hamiltonain 

\begin{align}
\mathcal{H} = J\sum_{\langle ij\rangle,\ell}{\bf S}_{i,\ell}\cdot {\bf S}_{j,\ell}+\sum_{\langle ij\rangle,\ell}{\bf D}_{ij}\cdot{\bf S}_{i,\ell}\times {\bf S}_{j,\ell}+J_c\sum_{\langle \ell\ell^\prime\rangle,i}{\bf S}_{i,\ell}\cdot {\bf S}_{i,\ell^\prime} - H\sum_i S_i^z,
\end{align}
 where $i$ and $j$ denote the sites on  the kagome layers, $\ell$ and $\ell^\prime$ label the layers, and $\langle ij\rangle$ denotes summation over nearest-neighbour.  The first term is an intralayer antiferromagnetic nearest-neighbour Heisenberg interaction denoted by $J$.  The second term is the Dzyaloshinskii-Moriya interaction  (DMI), where ${\bf D}_{ij}$ is the DM vector between site $i$ and $j$,  due to lack of inversion symmetry between two sites on each kagom\'e  layer. It is a perturbative anisotropy to the Heisenberg interaction $J$, and it is dominated by the out-of-plane component  (i.e. ${\bf D}_{ij}=\pm D_z{\bf \hat z}$). 
 
The DMI alternates between the triangular plaquettes of the kagom\'e lattice. The out-of-plane DMI stabilizes  the $120^\circ$ non-collinear spin structure and its sign  determines the vector chirality of the non-collinear spin order.  The third term  is an interlayer NN  Heisenberg interaction  between the kagome  layers denoted by $J_c$. It can either be  ferromagnetic ($J_c<0$) or antiferromagnetic ($J_c>0$). The last term  is an external magnetic field with strength $H$ in units of $g\mu_B$, and it is applied along the stacking direction taken as the $z$-axis. Without loss of generality, we consider unshifted  stacked kagome layers as  realized in different stacked frustrated kagome antiferromagnets.   The rigorous mathematical details of the  analysis can be found in the paper referenced above.
