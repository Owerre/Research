# Weyl Magnons in Stacked Kagome Antiferromagnets

In this notebook, I present the python code for one of my research papers, [Phys. Rev. B 97, 094412 (2018)](https://doi.org/10.1103/PhysRevB.97.094412). This paper proposed the first existence of Weyl magnons in stacked kagome antiferromagnets, governed by the Heisenberg Hamiltonain 

\begin{align}
\mathcal{H} = J\sum_{\langle ij\rangle,\ell}{\bf S}_{i,\ell}\cdot {\bf S}_{j,\ell}+\sum_{\langle ij\rangle,\ell}{\bf D}_{ij}\cdot{\bf S}_{i,\ell}\times {\bf S}_{j,\ell}+J_c\sum_{\langle \ell\ell^\prime\rangle,i}{\bf S}_{i,\ell}\cdot {\bf S}_{i,\ell^\prime} - H\sum_i S_i^z,
\end{align}
 where $i$ and $j$ denote the sites on  the kagome layers, $\ell$ and $\ell^\prime$ label the layers, and $\langle ij\rangle$ denotes summation over nearest-neighbour.  The first term is an intralayer antiferromagnetic nearest-neighbour Heisenberg interaction denoted by $J$.  The second term is the Dzyaloshinskii-Moriya interaction  (DMI), where ${\bf D}_{ij}$ is the DM vector between site $i$ and $j$,  due to lack of inversion symmetry between two sites on each kagom\'e  layer. It is a perturbative anisotropy to the Heisenberg interaction $J$, and it is dominated by the out-of-plane component  (i.e. ${\bf D}_{ij}=\pm D_z{\bf \hat z}$). 
 
The DMI alternates between the triangular plaquettes of the kagome lattice. The out-of-plane DMI stabilizes  the $120^\circ$ non-collinear spin structure and its sign  determines the vector chirality of the non-collinear spin order.  The third term  is an interlayer NN  Heisenberg interaction  between the kagome  layers denoted by $J_c$. It can either be  ferromagnetic ($J_c<0$) or antiferromagnetic ($J_c>0$). The last term  is an external magnetic field with strength $H$ in units of $g\mu_B$, and it is applied along the stacking direction taken as the $z$-axis. Without loss of generality, we consider unshifted  stacked kagome layers as  realized in different stacked frustrated kagome antiferromagnets.   The rigorous mathematical details of the  analysis can be found in the paper referenced above.

## 3. Berry curvature

```python
steps = 30 # size of array
Hsize = 6; # hamiltonian size
ky = np.sqrt(3) # fixed momentum vector

# initialize empty arrays
Ene = zeros((steps, steps, Hsize))
Omega = zeros((steps, steps, Hsize))

# mesh grid
kx = linspace(-1,1,steps)
kz = linspace(-1,1,steps)
Kx, Kz = meshgrid(kx, kz)

# compute Berry curvature and gradient
BC = Weyl.berry_curvature_kxkz(pi*kx, ky, pi*kz)
Px,Pz = gradient(BC[0])
Pxr = Px/sqrt(Px**2 +Pz**2)
Pzr = Pz/sqrt(Px**2 +Pz**2)

# plot figure
fig, ax = plt.subplots(figsize=(8,6))
ax.quiver(Kx,Kz,Pxr,Pzr, color = 'b', linewidth = 5)
ax.set_xlabel(r'$k_x/\pi$')
ax.set_ylabel(r'$k_z/\pi$')
ax.set_title("Monopole distribution of the lowest band")
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.xlim([-1, 1])
plt.ylim(-1, 1)
plt.savefig('../images/fig6b.png')
```

## 4. Chern numbers

```python
# set parameters
steps = 100
k1_int = 0
k1_fin = pi
Dk1 = (k1_fin - k1_int) / steps

k2_int = 0
k2_fin = pi
Dk2 = (k2_fin - k2_int) / steps

kz = linspace(-1,1,steps)

Nd = 6  # dimension of the Hamiltonian

# initialize arrays
chernN = np.zeros((Nd,len(kz)), dtype=complex)
sumN = np.zeros((Nd), dtype=complex)

for k in range(steps):
    for i in range(steps):
        k1 = k1_int + i * Dk1
        for j in range(steps):
            k2 = k2_int + j * Dk2
            k_vec = np.array([k1, k2, pi*kz[k]], float)
            Dk = np.array([Dk1, Dk2], float)
            LF = Weyl.latF_xy(k_vec, Dk, Nd)
            sumN += LF
            # LF_arr[:, i, j] = LF.imag # save data for plotting
    chernN[:,k] = sumN.imag / (2 *np.pi)
    sumN = np.zeros((Nd), dtype=complex)
# print("Chern numbers = {}".format(chernN))
# print("Sum of Chern Number = {} ".format(sum(chernN)))

# plot figure
plt.figure(figsize = (20,8))
plt.subplot(121)
plt.plot(kz,chernN[3,:], color = 'b')
plt.xlabel(r'$k_z/\pi$')
plt.ylabel(r'Chern number')
plt.xticks([-1,-0.5,0,0.5,1])
plt.yticks([-1,0,1])
plt.title('Chern number of the lowest Weyl band')

plt.subplot(122)
plt.plot(kz,chernN[4,:], color = 'b')
plt.xlabel(r'$k_z/\pi$')
plt.ylabel(r'Chern number')
plt.xticks([-1,-0.5,0,0.5,1])
plt.yticks([-2,-1,0])
plt.title('Chern number of the second Weyl band')
plt.savefig('../images/fig7.png')
```
