# Magnonic Floquet Spin Hall  Insulator In Bilayer Collinear Antiferromagnets

In this notebook, I present the python code for one of my research papers [Scientific Reports 9, 7197 (2019)](https://doi.org/10.1038/s41598-019-43702-9). In this paper, studied  bilayer Heisenberg Hamiltonain on the honeycomb lattice given by

\begin{align}
 \mathcal{H} = J\sum_{\langle ij\rangle,\ell}{\bf S}_{i,\ell}\cdot {\bf S}_{j,\ell}+J_c\sum_{i}{\bf S}_{i}^T\cdot {\bf S}_{i}^B,
 \end{align}

where ${\bf S}_{i}$ is the spin vector at site  $i$ and $\ell$ labels the top ($T$) and bottom ($B$) layers. The first term is the intralayer coupling in each layer and $\langle ij\rangle$ denotes summation over nearest neighbor spin sites and the second term is the interlayer coupling between the spins on the top and bottom layer.  I will consider two different cases: $(i)$ $J>0, J_c >0$  and $(ii)$ $J<0, J_c >0$.  The intralayer and interlayer couplings are antiferromagnetic in case $(i)$, whereas in case $(ii)$ the intralayer coupling is antiferromagnetic and the interlayer coupling is ferromagnetic. In both cases  the net magnetization vanishes. We note that case $(ii)$ is manifested in the  bilayer honeycomb magnet CrI$_3$.  The rigorous mathematical details of the  analysis can be found in the paper referenced above.
